```python
import requests
import pandas as pd
import io
urls = {
    "listings": "listings",
    "calendar": "calendar",
    "reviews":  "reviews"
}
for name, url in urls.items():
    print(f"\n--- {name.upper()} ---")
    r = requests.get(url)
    r.raise_for_status()
    df = pd.read_csv(io.BytesIO(r.content), compression="gzip", low_memory=False)
    print(list(df.columns))
```
